```python
import pandas as pd
import os

pd.options.display.max_rows = 150
```

```python
search_df_nm = '포괄손익계산서'

data_list = os.listdir()

match_list = list()

for df_nm in data_list:
    if search_df_nm in df_nm:
        match_list.append(df_nm)
        
match_list     
```

```python
df_list = list()

for i, df_nm in enumerate(match_list):
    df = pd.read_csv('./{}'.format(df_nm), encoding='utf-8-sig')
    df = df.iloc[:,1:3].copy()
    df = df[df['과 목'].notnull()]
    df['과 목'] = df['과 목'].str.strip()
    df['과 목'] = df['과 목'].str.replace(' ','')
    df['과 목'] = df['과 목'].str.split('.').str[-1]
    df['과 목'] = df['과 목'].str.strip()
    df['과 목'] = df['과 목'].str.split('(').str[0].copy()
    df.rename(columns = {'과 목' : '계정과목'}, inplace = True)
    df_list.append(df)
```

```python
df_list_index = list()

for i, val in enumerate(df_list):
    df_index = df_list[i].set_index('계정과목')
    df_list_index.append(df_index)
```

```python
df_merge = pd.merge(df_list_index[0], df_list_index[1], how='outer', left_index=True, right_index=True )
```

```python
df_merge = pd.merge(df_merge, df_list_index[2], how='outer', left_index=True, right_index=True )
```

```python
df_merge = pd.merge(df_merge, df_list_index[3], left_index = True, right_index=True, how='outer')
```

```python
df_merge = pd.merge(df_merge, df_list_index[4], left_index = True, right_index=True, how='outer')
df_merge
```
