```python
path = 'COVID-19-master/csse_covid_19_data/csse_covid_19_daily_reports/'

with open('COVID-19-master/csse_covid_19_data/country_convert.json', 'r', encoding = 'utf-8-sig') as json_file:
    json_data = json.load(json_file)

def country_name_change(df):
    if df['Country_Region'] in json_data:
        df['Country_Region'] = json_data[df['Country_Region']]
    return df['Country_Region']

def create_df(data):
    df = pd.read_csv(path + data, encoding = 'utf-8-sig')
    try:
        df = df[['Country_Region','Deaths']]
    except:
        df = df[['Country/Region','Deaths']]
        df.columns = df.columns.str.replace('/','_')

    df = df.dropna(subset=['Deaths'])
    df['Country_Region'] = df.apply(country_name_change, axis = 1)
    df.astype({'Deaths':'int64'})
    df = df.groupby('Country_Region').sum()

    dates = data.split('.')[0].lstrip('0').replace('-','/')
    df.columns = [dates]

    return df
```

```python
def merge_df(path):
    
    file_list, csv_list = os.listdir(path), list()
    df_first = True
    

    for file in file_list:
        if file.split('.')[-1] == 'csv':
            csv_list.append(file)
    csv_list.sort()
    
    for file in csv_list:
        df = create_df(file)
        if df_first:
            df_final, df_first = df, False
        else:
            df_final = pd.merge(df_final, df, how = 'outer', left_index = True, right_index  =True)

    df_final = df_final.fillna(0)
    return df_final
```

```python
path = 'COVID-19-master/csse_covid_19_data/csse_covid_19_daily_reports/'
df = merge_df(path)
df
```

```python
df = df.astype('int64')
df
```
