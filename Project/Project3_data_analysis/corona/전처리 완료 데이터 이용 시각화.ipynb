```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('전처리 완료.csv', encoding='utf-8-sig')
```

```python
df.head()
```

```python
df_iso = pd.read_csv('COVID-19-master/csse_covid_19_data/UID_ISO_FIPS_LookUp_Table.csv', encoding = 'utf-8-sig', keep_default_na=False, na_values='')
df_iso.head()
```

```python
df_iso = df_iso[['iso2', 'Country_Region']]
```

```python
df_iso = df_iso.drop_duplicates(subset = 'Country_Region', keep='last')
```

```python
df_merge = pd.merge(df, df_iso, how = 'left', on = 'Country_Region')
```

```python
df = df_merge.dropna(subset = ['iso2'])
```

```python
def flag(row):
    flag_link = 'https://www.countryflags.io/' + row + '/flat/64.png'
    return flag_link

df['iso2'] = df['iso2'].apply(flag)
```

```python
df
```

```python
cols = df.columns.tolist()
cols.remove('iso2')
cols.insert(1,'iso2')
df = df[cols]
```

```python
df
```

```python
cols[1] = 'country_flag'
```

```python
df.columns = cols
```
